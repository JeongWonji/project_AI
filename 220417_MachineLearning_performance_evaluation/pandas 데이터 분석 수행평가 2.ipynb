```python
import numpy as np
import pandas as pd

a = pd.read_csv('./data/ta_20220321160757(서울).csv', encoding='CP949', sep='\t')
b = pd.read_csv('./data/ta_20220321160847(대구).csv', encoding='cp949', sep='\t')

display(a)

display(b)
```

```python
seoul = []

# 날짜 지점 평균기온 최저기온 최고기온

for i in range(len(a)):
    seoul.append(a.values[i][0].split(','))

df1 = pd.DataFrame(np.array(seoul),
                  columns = list('12345'))
df1.columns = ['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
df1
```

```python
daegu = []

# 날짜 지점 평균기온 최저기온 최고기온

for i in range(len(b)):
    daegu.append(b.values[i][0].split(','))

df2 = pd.DataFrame(np.array(daegu),
                  columns = list('12345'))
df2.columns = ['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
df2
```
