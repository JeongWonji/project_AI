```python
# 3.

import numpy as np
import pandas as pd

df1 = pd.read_csv('./data/ratings.csv')
df2 = pd.read_csv('./data/movies.csv')


df3 = df1.groupby(['movieId']).mean()

df = pd.merge(left = df2 , right = df3, how = "inner", on = "movieId")

df4 =df.loc[:, ['movieId', 'title', 'rating']] 

display(df4)
```

```python
# 4.

df5 = df4[df4['rating'] == 5.0].sort_values('title', ascending = True)

display(df5)
```
